# 06 — HTML5 Structural Validation

The HTML5 schema enforces W3C nesting rules.

**What you learn:**
- HtmlBuilder validates element nesting at insertion time
- Invalid nesting raises `ValueError` immediately

**Prerequisites:** 05_components

```python
from IPython.display import HTML
from genro_toolbox import smartawait
from genro_builders.contrib.html import HtmlBuilder
```

```python
builder = HtmlBuilder()
s = builder.source
s.head().style('body{font-family:sans-serif;max-width:600px;margin:2em auto} table{border-collapse:collapse;width:100%;margin:1em 0} th,td{padding:.5em;border:1px solid #ddd;text-align:left} th{background:#f0f4f8}')

body = s.body()
body.h1('Structural Validation')
body.p('Valid structures:')

ul = body.ul()
ul.li('Item 1')
ul.li('Item 2')

table = body.table()
thead = table.thead()
row = thead.tr()
row.th('Name')
row.th('Value')
tbody = table.tbody()
r = tbody.tr()
r.td('Alpha')
r.td('100')

await smartawait(builder.build())
HTML(builder.render())
```

## Invalid nesting — caught at insertion time

```python
cases = [
    ('div > tr',  lambda: HtmlBuilder().source.body().div().tr()),
    ('tbody > td', lambda: HtmlBuilder().source.body().table().tbody().td('x')),
    ('div > li',  lambda: HtmlBuilder().source.body().div().li('x')),
    ('p > div',   lambda: HtmlBuilder().source.body().p().div()),
]

for label, fn in cases:
    try:
        fn()
        print(f'{label}: no error (unexpected)')
    except ValueError as e:
        print(f'{label}: {e}')
```
