# 03 — BuilderManager

Separating data from structure with the manager lifecycle.

**What you learn:**
- `BuilderManager` coordinates builders with a shared data store
- `store(data)`: populate shared data (called first)
- `main(source)`: build the HTML structure (called second)
- `run()` returns an awaitable in async context (Jupyter)

**Prerequisites:** 02_static_page

```python
from IPython.display import HTML
from genro_toolbox import smartawait
from genro_builders.contrib.html import HtmlBuilder
from genro_builders.manager import BuilderManager
```

```python
class ContactList(BuilderManager):
    def __init__(self):
        self.page = self.set_builder('page', HtmlBuilder)

    def store(self, data):
        data['contacts'] = [
            ('Alice Johnson', 'alice@example.com', 'Engineer'),
            ('Bob Smith', 'bob@example.com', 'Designer'),
            ('Carol White', 'carol@example.com', 'Manager'),
        ]

    def main(self, source):
        source.head().style('body{font-family:sans-serif;max-width:600px;margin:2em auto} table{width:100%;border-collapse:collapse} th,td{padding:.5em;text-align:left;border-bottom:1px solid #ddd} th{background:#f0f4f8}')
        body = source.body()
        body.h1('Team Contacts')
        table = body.table()
        thead = table.thead()
        header = thead.tr()
        header.th('Name')
        header.th('Email')
        header.th('Role')
        tbody = table.tbody()
        for name, email, role in self.reactive_store['contacts']:
            row = tbody.tr()
            row.td(name)
            row.td(email)
            row.td(role)
```

```python
app = ContactList()
await smartawait(app.run())
HTML(app.page.render())
```
