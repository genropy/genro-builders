# 03 — HtmlManager

Separating data from structure with the manager lifecycle.

**What you learn:**
- `HtmlManager` coordinates a builder with a shared data store
- `store(data)`: populate shared data (called first)
- `main(source)`: build the HTML structure (called second)
- `render()`: auto-runs if needed, produces HTML

**Prerequisites:** 02_static_page

```python
from IPython.display import HTML

from genro_builders.contrib.html import HtmlManager
```

```python
class ContactList(HtmlManager):
    def store(self, data):
        data['contacts'] = [
            ('Alice Johnson', 'alice@example.com', 'Engineer'),
            ('Bob Smith', 'bob@example.com', 'Designer'),
            ('Carol White', 'carol@example.com', 'Manager'),
        ]

    def main(self, source):
        source.head().style("""
            .page { font-family: sans-serif; max-width: 600px; margin: 2em auto;
                    color: #333; background: #fff; padding: 1.5em; border-radius: 8px; }
            table { width: 100%; border-collapse: collapse; }
            th, td { padding: 0.5em; text-align: left; border-bottom: 1px solid #ddd; }
            th { background: #f0f4f8; color: #1e293b; }
            h1 { color: #1e293b; }
        """)
        body = source.body()
        page = body.div(_class='page')
        page.h1('Team Contacts')
        table = page.table()
        thead = table.thead()
        header = thead.tr()
        header.th('Name')
        header.th('Email')
        header.th('Role')
        tbody = table.tbody()
        for name, email, role in self.reactive_store['contacts']:
            row = tbody.tr()
            row.td(name)
            row.td(email)
            row.td(role)
```

```python
app = ContactList()
HTML(app.render())
```
