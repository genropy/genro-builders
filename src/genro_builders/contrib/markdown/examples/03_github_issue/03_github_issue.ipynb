# 03 — GitHub Issue Template

Build a structured GitHub issue template using `BuilderManager`.

**What you learn:**
- Use `BuilderManager` to separate data from structure
- `store()` method to define initial data
- `main()` method to define the document structure
- Generate a bug report template from data

**Prerequisites:** 02_document_structure

```python
from IPython.display import Markdown
from genro_builders.contrib.markdown import MarkdownBuilder
from genro_builders.manager import BuilderManager


class BugReport(BuilderManager):
    def __init__(self):
        self.doc = self.set_builder('doc', MarkdownBuilder)
        self.run()

    def store(self, data):
        data['title'] = 'Login fails with SSO on Safari'
        data['labels'] = 'bug, auth, priority:high'
        data['assignee'] = 'gporcari'
        data['environment'] = 'macOS 15.3 / Safari 18.2'
        data['description'] = (
            'When attempting to log in via SSO (SAML), Safari '
            'redirects to a blank page instead of completing the auth flow.'
        )
        data['steps'] = [
            'Go to /login',
            'Click "Sign in with SSO"',
            'Enter company email and submit',
            'Observe blank page after IdP redirect',
        ]
        data['expected'] = 'User is redirected to the dashboard after successful authentication.'
        data['actual'] = 'Safari shows a blank page. Console shows a CORS error on the callback URL.'

    def main(self, source):
        data = self.reactive_store

        source.h1(f"[BUG] {data['title']}")

        source.h2("Metadata")
        meta = source.table()
        for label, key in [('Labels', 'labels'), ('Assignee', 'assignee'), ('Environment', 'environment')]:
            row = meta.tr()
            row.th(label)
            row.td(str(data[key]))

        source.h2("Description")
        source.p(str(data['description']))

        source.h2("Steps to Reproduce")
        steps = source.ol()
        for step in data['steps']:
            steps.li(str(step))

        source.h2("Expected Behavior")
        source.p(str(data['expected']))

        source.h2("Actual Behavior")
        source.p(str(data['actual']))
```

```python
app = BugReport()
md = app.doc.render()
print(md)
```
