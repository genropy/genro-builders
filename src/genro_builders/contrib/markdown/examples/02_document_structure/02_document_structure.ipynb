# 02 — Document Structure

A complete tour of every MarkdownBuilder element.

**What you learn:**
- Headings (`h1`–`h6`)
- Block elements: `p`, `code`, `blockquote`, `hr`
- Tables: `table`, `tr`, `th`, `td`
- Lists: `ul`, `ol`, `li`
- Inline elements: `bold`, `italic`, `inlinecode`, `link`, `img`, `text`

**Prerequisites:** 01_hello_world

```python
from genro_builders.contrib.markdown import MarkdownBuilder
from IPython.display import Markdown

builder = MarkdownBuilder()
doc = builder.source
```

## Headings and Paragraphs

```python
doc.h1("Project Documentation")
doc.p("This document showcases every element available in MarkdownBuilder.")

doc.h2("Getting Started")
doc.p("Install the package and import the builder.")
```

## Code Blocks

```python
doc.h2("Installation")
doc.code("pip install genro-builders", lang="bash")

doc.h2("Quick Example")
doc.code(
    "from genro_builders.contrib.markdown import MarkdownBuilder\n"
    "builder = MarkdownBuilder()\n"
    "builder.source.h1('Hello')\n"
    "builder.build()\n"
    "print(builder.render())",
    lang="python",
)
```

## Blockquote and Horizontal Rule

```python
doc.blockquote("The best way to predict the future is to invent it.\n— Alan Kay")
doc.hr()
```

## Tables

```python
doc.h2("Comparison")
table = doc.table()

header = table.tr()
header.th("Feature")
header.th("MarkdownBuilder")
header.th("Manual String")

for feature, builder_val, manual_val in [
    ("Validation", "Yes", "No"),
    ("Composable", "Yes", "No"),
    ("Reactive", "Yes", "No"),
]:
    row = table.tr()
    row.td(feature)
    row.td(builder_val)
    row.td(manual_val)
```

## Lists

```python
doc.h2("Features")
features = doc.ul()
features.li("Grammar-based validation")
features.li("Declarative rendering")
features.li("Reactive data binding")

doc.h2("Roadmap")
roadmap = doc.ol()
roadmap.li("Core builder")
roadmap.li("Renderer")
roadmap.li("Compiler")
```

## Inline Elements

```python
doc.h2("Resources")
doc.link("Documentation", href="https://genro.dev/docs")
doc.link("GitHub", href="https://github.com/softwell/genro-builders")
doc.p("Use **bold** for emphasis and `code` for inline snippets.")
doc.bold("Important note")
doc.italic("Subtle remark")
doc.inlinecode("builder.render()")
```

## Build and Render

```python
builder.build()
md = builder.render()
print(md)
```

## Preview

```python
Markdown(md)
```
