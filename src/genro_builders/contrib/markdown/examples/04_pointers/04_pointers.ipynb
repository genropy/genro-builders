# 04 — Pointers

Declarative data binding with `^pointer` syntax.

**What you learn:**
- `'^key'` as first arg binds element content to data
- Data is resolved at render time from the reactive store
- Changing data and rebuilding produces different output
- Same template, different issue — just change the data

**Prerequisites:** 03_github_issue

```python
from IPython.display import Markdown
from genro_builders.contrib.markdown import MarkdownBuilder
from genro_builders.manager import BuilderManager


class IssueTemplate(BuilderManager):
    """A reusable issue template where content comes from data, not code."""

    def __init__(self):
        self.doc = self.set_builder('doc', MarkdownBuilder)
        self.run()

    def store(self, data):
        data['title'] = 'API returns 500 on empty payload'
        data['label'] = 'bug'
        data['assignee'] = 'gporcari'
        data['description'] = (
            'POST /api/v1/items with an empty JSON body '
            'returns 500 instead of 422 Unprocessable Entity.'
        )
        data['expected'] = '422 response with validation error details.'
        data['actual'] = '500 Internal Server Error with a NoneType traceback.'

    def main(self, source):
        # ^pointer: content resolved from reactive_store at render time
        source.h1('^title')

        source.h2('Metadata')
        meta = source.table()
        row_label = meta.tr()
        row_label.th('Label')
        row_label.td('^label')
        row_assignee = meta.tr()
        row_assignee.th('Assignee')
        row_assignee.td('^assignee')

        source.h2('Description')
        source.p('^description')

        source.h2('Expected Behavior')
        source.p('^expected')

        source.h2('Actual Behavior')
        source.p('^actual')
```

```python
app = IssueTemplate()
Markdown(app.doc.render())
```

## Change data and rebuild

Same template structure — completely different issue. Just swap the data:

```python
app.reactive_store['title'] = 'Dark mode toggle not persisted'
app.reactive_store['label'] = 'enhancement'
app.reactive_store['assignee'] = 'mrossi'
app.reactive_store['description'] = (
    'User enables dark mode in Settings, but after page refresh '
    'the preference is lost and light mode is restored.'
)
app.reactive_store['expected'] = 'Dark mode preference is saved to localStorage and restored on load.'
app.reactive_store['actual'] = 'Preference is only in memory — lost on refresh.'

app.doc.build()
Markdown(app.doc.render())
```
