# 04 — Pointers

Declarative data binding with `^pointer` syntax.

**What you learn:**
- `'^key'` as first arg binds element content to data
- Data is resolved at render time from the reactive store
- Changing data and rebuilding produces different output
- Same template, different issue — just change the data

**Prerequisites:** 03_github_issue

```python
from pathlib import Path
from IPython.display import Markdown

from genro_bag import Bag
from genro_builders.contrib.markdown import MarkdownBuilder
from genro_builders.manager import BuilderManager
from genro_toolbox import set_sync

set_sync()


class IssueTemplate(BuilderManager):
    """A reusable issue template where content comes from data, not code."""

    def on_init(self):
        self.doc = self.register_builder('doc', MarkdownBuilder)
        self.run()

    def main(self, source):
        source.h1('^title')

        source.h2('Metadata')
        meta = source.table()
        row_label = meta.tr()
        row_label.th('Label')
        row_label.td('^label')
        row_assignee = meta.tr()
        row_assignee.th('Assignee')
        row_assignee.td('^assignee')

        source.h2('Description')
        source.p('^description')

        source.h2('Expected Behavior')
        source.p('^expected')

        source.h2('Actual Behavior')
        source.p('^actual')
```

```python
app = IssueTemplate()
app.local_store('doc').fill_from(
    Bag.from_json(Path('issue.json').read_text()),
)
app.doc.build()
Markdown(app.doc.render())
```

## Swap data file — same template, different issue

```python
app2 = IssueTemplate()
app2.local_store('doc').fill_from(
    Bag.from_json(Path('issue_alt.json').read_text()),
)
app2.doc.build()
Markdown(app2.doc.render())
```
