# 01 — Hello World

The simplest possible Markdown builder: create a document and render it.

**What you learn:**
- Instantiate `MarkdownBuilder`
- Access the source Bag via `builder.source`
- Add elements (`h1`, `p`)
- `build()` materializes source → built
- `render()` produces the Markdown string

**Prerequisites:** None — this is the starting point.

```python
from genro_builders.contrib.markdown import MarkdownBuilder

# Create a builder — this gives us the Markdown grammar
builder = MarkdownBuilder()

# builder.source is where we describe the structure
builder.source.h1("Hello World")
builder.source.p("This is my first Markdown builder document.")
```

## Build and Render

`build()` expands the source into the built Bag. `render()` serializes it to Markdown.

```python
# build() materializes source → built
builder.build()

# render() produces the Markdown string
md = builder.render()
print(md)
```

## Preview

Render the Markdown inline in the notebook:

```python
from IPython.display import Markdown

Markdown(md)
```
