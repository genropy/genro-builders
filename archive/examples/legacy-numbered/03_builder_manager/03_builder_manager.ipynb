# 03 — HtmlManager

Separating data from structure with the manager lifecycle.

**What you learn:**
- `HtmlManager` coordinates a builder with a shared data store
- Data lives in external JSON files, loaded at runtime
- CSS lives in external files
- `main(source)`: defines only the HTML structure

**Prerequisites:** 02_static_page

```python
from IPython.display import HTML

from genro_bag.resolvers import FileResolver
from genro_builders.contrib.html import HtmlManager
from genro_toolbox import set_sync

set_sync()
```

```python
class ContactList(HtmlManager):
    def main(self, source):
        # FileResolver loads style.css lazily at render time (pull model)
        source.head().style(FileResolver('style.css'))

        body = source.body()
        page = body.div(_class='page')
        page.h1('Team Contacts')

        contacts = self.local_store()['contacts']

        table = page.table()
        thead = table.thead()
        header = thead.tr()
        header.th('Name')
        header.th('Email')
        header.th('Role')
        tbody = table.tbody()
        for contact in contacts:
            row = tbody.tr()
            row.td(contact.attr.get('name'))
            row.td(contact.attr.get('email'))
            row.td(contact.attr.get('role'))
```

```python
app = ContactList()
# FileResolver with as_bag=True: Bag node format (label/value/attr) preserved
app.local_store('page').set_resolver(
    'contacts', FileResolver('contacts.json', as_bag=True),
)
HTML(app.render())
```
