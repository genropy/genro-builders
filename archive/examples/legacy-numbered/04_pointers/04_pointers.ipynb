# 04 — Pointers

Declarative data binding with `^pointer` syntax.

**What you learn:**
- `'^key'` as first arg binds element content to data
- Data is resolved at render time from the builder's namespace
- Changing data produces different output on next render

**Prerequisites:** 03_builder_manager

```python
from pathlib import Path
from IPython.display import HTML

from genro_bag import Bag
from genro_bag.resolvers import FileResolver
from genro_builders.contrib.html import HtmlManager
from genro_toolbox import set_sync

set_sync()
```

```python
class ProductCard(HtmlManager):
    def main(self, source):
        # FileResolver loads style.css lazily at render time (pull model)
        source.head().style(FileResolver('style.css'))
        body = source.body()
        card = body.div(_class='page')
        # ^pointer: content resolved from builder's namespace at render time
        card.h2('^name')
        card.p('^price', _class='price')
        card.p('^description')
        card.p('^in_stock', _class='stock')
```

```python
app = ProductCard()
app.local_store('page').fill_from(
    Bag.from_json(Path('product.json').read_text()),
)
HTML(app.render())
```

## Swap data file — same structure, different content

```python
app2 = ProductCard()
app2.local_store('page').fill_from(
    Bag.from_json(Path('product_alt.json').read_text()),
)
HTML(app2.render())
```
