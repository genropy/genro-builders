# 01 — Hello World

The simplest possible builder app: subclass `HtmlManager`, implement `main()`, render.

**What you learn:**
- `HtmlManager`: the standard entry point for HTML apps
- `main(source)`: populate the source Bag with elements
- `render()`: produce the HTML string (auto-builds if needed)

**Prerequisites:** None.

```python
from IPython.display import HTML

from genro_builders.contrib.html import HtmlManager
```

```python
class HelloWorld(HtmlManager):
    def main(self, source):
        body = source.body()
        body.h1('Hello World')
        body.p('This is my first builder page.')
```

```python
app = HelloWorld()
html = app.render()
print(html)
```
