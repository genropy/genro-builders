# 02 — Static Page

Structure, nesting, and **composition**: split `main()` into methods.

**What you learn:**
- Split `main()` into logical methods (head, hero, navigation, features)
- Full HTML structure: `head` + `body`
- Nesting: `div > h2 + ul > li`
- Attributes: `_class` (maps to `class`), `id`, `href`
- Python methods = reusable building blocks
- `FileResolver` (from genro-bag): load CSS from external file lazily

**Genro ecosystem: resolvers**

A **resolver** is a pull-based mechanism from genro-bag. Instead of storing
a value directly, you install a resolver that computes/loads the value
on first access. `FileResolver('style.css')` reads the file at render
time — if the file changes, the next render picks up the new version.
Same pull model used by `data_formula` for computed values.

**Prerequisites:** 01_hello_world

```python
from IPython.display import HTML

from genro_bag.resolvers import FileResolver
from genro_builders.contrib.html import HtmlManager
from genro_toolbox import set_sync

# Jupyter has an event loop, but we operate sync — accept blocking I/O
set_sync()
```

```python
class StaticPage(HtmlManager):
    """A static page composed from logical building blocks."""

    def main(self, source):
        self.page_head(source)
        body = source.body()
        page = body.div(_class='page')
        self.hero(page)
        self.navigation(page)
        self.features(page)
        page.footer().p('Built with genro-builders.')

    def page_head(self, source):
        head = source.head()
        head.meta(charset='utf-8')
        head.title('My Static Page')
        # FileResolver reads style.css lazily at render time.
        # The file is NOT read here — only when the HTML is produced.
        head.style(FileResolver('style.css'))

    def hero(self, page):
        hero = page.div(_class='hero')
        hero.h1('Welcome to GenroBuilders')
        hero.p('A declarative way to build HTML, SVG, Markdown, and more.')

    def navigation(self, page):
        nav = page.nav()
        ul = nav.ul(_class='nav')
        ul.li().a('Home', href='#home')
        ul.li().a('Docs', href='#docs')
        ul.li().a('Examples', href='#examples')

    def features(self, page):
        content = page.div(id='content')
        content.h2('Features')
        features = content.ul()
        features.li('112 HTML5 elements with grammar validation')
        features.li('Declarative data binding with ^pointers')
        features.li('Reactive updates via subscribe()')
        features.li('Components for reusable UI patterns')
```

```python
app = StaticPage()
HTML(app.render())
```
